Simple Linear Regression is a statistical model that estimates the relationship between two continuous variable a predictor and an outcome, they are two continuous variables. Outcome variable is the dependent variable we aim to predict or explain. Predict variable is the independent variable used to explain changes in the outcome variable. Slope represents the change in outcome variable for each one-unit increase in predict variable. Intercept coefficient represents the change in outcome variable for each one-unit increase in predict variable. Error term represents the random deviation of the observed values from the predicted values. 

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set random seed for reproducibility
np.random.seed(42)

# Define model parameters
beta_0 = 5    # Intercept
beta_1 = 2    # Slope
sigma = 1     # Standard deviation of the error term

# Generate predictor variable (X)
X = np.linspace(0, 10, 100)  # 100 points between 0 and 10

# Generate error term (epsilon) from a normal distribution
epsilon = np.random.normal(0, sigma, size=X.shape)

# Calculate the outcome variable (Y) using the linear model
Y = beta_0 + beta_1 * X + epsilon

# Plot the data and the true regression line
plt.scatter(X, Y, color='blue', label='Observed Data')
plt.plot(X, beta_0 + beta_1 * X, color='red', label='True Regression Line')
plt.xlabel('Predictor Variable (X)')
plt.ylabel('Outcome Variable (Y)')
plt.title('Simple Linear Regression: Y = 0 + 1X + ')
plt.legend()
plt.show()
````

# summary

1. **Components of the Model**:
   - **Outcome Variable (Y)**: The dependent variable we aim to predict.
   - **Predictor Variable (X)**: The independent variable used to explain changes in \( Y \).
   - **Slope Coefficient (\(\beta_1\))**: Shows the change in \( Y \) per unit increase in \( X \).
   - **Intercept (\(\beta_0\))**: The expected \( Y \) value when \( X = 0 \).
   - **Error Term (\(\epsilon\))**: A random deviation that follows a normal distribution with mean zero, reflecting the difference between observed and predicted values.

2. **Model Equation**:
   The equation for Simple Linear Regression is:
   \[
   Y = \beta_0 + \beta_1 X + \epsilon
   \]
   This equation suggests that \( Y \) can be understood as a combination of the linear relationship (\( \beta_0 + \beta_1 X \)) and some random variation (error term \( \epsilon \)).

3. **Normal Distribution Assumption**:
   In this model, \( Y \) is assumed to follow a normal distribution with a mean given by \( \beta_0 + \beta_1 X \) and a standard deviation \( \sigma \) defined by the variability of \( \epsilon \).

4. **Python Simulation**:
   We wrote Python code to simulate a dataset following this model using `numpy` and `scipy.stats`. The code generated random values for \( X \) and added random error to produce \( Y \), displaying the observed data and the true regression line.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Step 1: Simulate the dataset
np.random.seed(42)
n = 100  # Number of data points
beta_0 = 5  # Intercept
beta_1 = 2  # Slope
sigma = 1   # Standard deviation of the error term

# Generate predictor variable (X) and error term (epsilon)
X = np.linspace(0, 10, n)
epsilon = np.random.normal(0, sigma, n)

# Calculate the outcome variable (Y) using the linear model
Y = beta_0 + beta_1 * X + epsilon

# Create a DataFrame for the simulated data
data = pd.DataFrame({'X': X, 'Y': Y})

# Step 2: Fit the Linear Regression model
model = smf.ols('Y ~ X', data=data).fit()

# Display a summary of the fitted model
print(model.summary())

# Step 3: Visualize the fitted model
# Plot observed data points
plt.scatter(data['X'], data['Y'], color='blue', label='Observed Data')

# Plot the fitted regression line
plt.plot(data['X'], model.predict(data['X']), color='red', label='Fitted Regression Line')

# Customize the plot
plt.xlabel('Predictor Variable (X)')
plt.ylabel('Outcome Variable (Y)')
plt.title('Fitted Simple Linear Regression Model')
plt.legend()
plt.show()
````

The theoretical line represents the idealized model, while the fitted line represents the model we estimated from the data. The fitted line is influenced by the random noise in the data, which accounts for the slight difference between the two. The theoretical line is fixed. Its the exact line we would expect in an ideal world without any random error. Since our observed data points deviate from the theoretical line due to this error, the fitted line will often differ slightly from the theoretical line. 

In the Simple Linear Regression context, we used Python code to explore the distinction between:

1. **Theoretical (True) Model**:
   - This is the model we used to generate the simulated data:
   \[
   Y_i = \beta_0 + \beta_1 x_i + \epsilon_i
   \]
   where \( \epsilon_i \sim \mathcal{N}(0, \sigma) \).
   - Here, \(Y_i\) is the true outcome for each \(x_i\), generated with the slope \( \beta_1 \), intercept \( \beta_0 \), and added random noise \( \epsilon_i \) to simulate real-world variability.

2. **Fitted (Estimated) Model**:
   - This is the model fitted to the simulated data, aiming to estimate the relationship between \(X\) and \(Y\):
   \[
   \hat{y}_i = \hat{\beta}_0 + \hat{\beta}_1 x_i
   \]
   where \(\hat{\beta}_0\) and \(\hat{\beta}_1\) are estimated from the data.
   - The fitted model represents the *best estimate* of the true relationship based on the observed data. It provides a regression line that may slightly differ from the theoretical line due to the random error in the data.

This distinction highlights that the fitted model may approximate, but not exactly match, the theoretical model due to the variability introduced by the random error term in the data.

In `statsmodels`, the fitted values from a regression model, available as `fitted_model.fittedvalues`, are derived using the estimated parameters (intercept and slope) from the regression output, specifically `fitted_model.params` or the table in `fitted_model.summary().tables[1]`.

Here's how it works in detail:

1. **Model Estimation and Parameters**:
   - When you fit a Simple Linear Regression model with `statsmodels`, it calculates the estimated intercept (\(\hat{\beta}_0\)) and slope (\(\hat{\beta}_1\)) for the predictor variable \(X\).
   - These estimates are stored in `fitted_model.params`, where:
     - `fitted_model.params['Intercept']` gives the estimated intercept \(\hat{\beta}_0\).
     - `fitted_model.params['X']` gives the estimated slope \(\hat{\beta}_1\).
   - These values are also available in the table shown in `fitted_model.summary().tables[1]`, under the columns "coef" (for the coefficient values) and "std err" (for their standard errors).

2. **Calculating Fitted Values**:
   - The fitted values for each observation are calculated using the regression equation:
     \[
     \hat{y}_i = \hat{\beta}_0 + \hat{\beta}_1 x_i
     \]
   - Here, \( \hat{\beta}_0 \) and \( \hat{\beta}_1 \) come from `fitted_model.params` or the summary table.
   - For each \(x_i\) in the data, the model multiplies \(x_i\) by \(\hat{\beta}_1\) and adds \(\hat{\beta}_0\), resulting in the predicted or fitted values stored in `fitted_model.fittedvalues`.

3. **Example Calculation**:
   Suppose we have the following estimated parameters from `fitted_model.params`:
   - Intercept (\(\hat{\beta}_0\)) = 5
   - Slope (\(\hat{\beta}_1\)) = 2

   Then, for a predictor value \(X = 3\):
   \[
   \hat{y} = 5 + 2 \times 3 = 11
   \]

   This fitted value, 11, would be one entry in `fitted_model.fittedvalues`.

4. **Python Code to Calculate Fitted Values Manually**:
   To illustrate, heres how you can manually compute fitted values using `fitted_model.params`:

   ```python
   # Manually calculate fitted values using the estimated parameters
   fitted_values_manual = fitted_model.params['Intercept'] + fitted_model.params['X'] * data['X']
   ```

   This manually computed `fitted_values_manual` would be identical to `fitted_model.fittedvalues`.

In summary, `fitted_model.fittedvalues` are derived by applying the estimated coefficients from `fitted_model.params` to the predictor values in the data. This process yields the predicted \( \hat{y} \) values, representing the best linear approximation based on the fitted model.

Heres a summary of our discussion on how `fitted_model.fittedvalues` are derived in a Simple Linear Regression model using `statsmodels`:

1. **Estimated Parameters**:
   - `fitted_model.params` contains the estimated intercept (\(\hat{\beta}_0\)) and slope (\(\hat{\beta}_1\)) values for the regression.
   - These parameter estimates are also available in `fitted_model.summary().tables[1]` under the "coef" column.

2. **Calculating Fitted Values**:
   - The fitted values (\(\hat{y}_i\)) for each observation are calculated as:
     \[
     \hat{y}_i = \hat{\beta}_0 + \hat{\beta}_1 x_i
     \]
   - Each \(x_i\) in the data is multiplied by the estimated slope, and the intercept is added, resulting in the fitted (predicted) values.

3. **Manual Calculation**:
   - We discussed how to manually calculate fitted values by using `fitted_model.params` and the datas predictor values, which yields the same results as `fitted_model.fittedvalues`.

In essence, `fitted_model.fittedvalues` provides the predicted \( \hat{y} \) values based on the estimated coefficients from the regression model. This process allows us to see how well the model's linear approximation aligns with the data.

In the Ordinary Least Squares (OLS) method, the fitted line chosen for the model is the one that minimizes the sum of the squared differences (errors) between the observed values and the predicted values. This line is selected to provide the best possible fit to the observed data.

The reason OLS uses **squares** of the errors is to ensure all errors are positive (avoiding cancellation between positive and negative errors) and to give more weight to larger errors, which penalizes large deviations more heavily. This squared-error minimization produces a single line that best represents the overall trend of the data by minimizing total error.

### 1. The first expression:
\[
1 - \frac{\sum (Y - \text{fitted_model.fittedvalues})^2}{\sum (Y - Y_{\text{mean}})^2}
\]

This expression represents **1 minus the ratio of the residual sum of squares to the total sum of squares**. Let's explain each term:

- **Residual Sum of Squares (RSS)**: This is the numerator of the fraction, \(\sum (Y - \text{fitted_model.fittedvalues})^2\). It captures the **unexplained variation** by the model  the squared differences between the actual \(Y\) values and the predicted \(Y\) values from the fitted model.
  
- **Total Sum of Squares (TSS)**: The denominator, \(\sum (Y - Y_{\text{mean}})^2\), captures the **total variation in the data** relative to the mean of \(Y\). It measures the spread of the \(Y\) values from their mean, without any consideration of the model.

The ratio of RSS/TSS quantifies how much of the total variation in \(Y\) is not explained by the model (i.e., how poorly the model fits the data). Subtracting this ratio from 1 gives us the proportion of the variation in \(Y\) that is **explained** by the model, which is commonly referred to as **R-squared**. Therefore, this expression essentially computes the **coefficient of determination**.

### 2. `fitted_model.rsquared`
This is a direct attribute of the fitted model that already calculates the R-squared value. It represents the proportion of the variance in the dependent variable \(Y\) that is explained by the model. The closer \(R^2\) is to 1, the better the model explains the variance in \(Y\). A value close to 0 would indicate that the model does not explain much of the variance in \(Y\).

### 3. `np.corrcoef(Y, fitted_model.fittedvalues)[0,1]**2`
This expression computes the **correlation coefficient** between the observed values \(Y\) and the fitted values (predicted values) from the model. The correlation coefficient is a measure of the linear relationship between \(Y\) and the fitted values:

- The **correlation coefficient** \(r\) quantifies how strongly \(Y\) and the fitted values are related, with values between \(-1\) and \(1\). Squaring this value (\(r^2\)) gives us the **proportion of variance** explained by the model, which is identical to \(R^2\).
  
Thus, this expression captures how much of the variation in \(Y\) can be explained by the fitted model.

### 4. `np.corrcoef(Y, x)[0,1]**2`
Here, we are looking at the **correlation between the observed values \(Y\) and the predictor variable \(x\)** (independent variable). Squaring this correlation coefficient gives the **proportion of variation in \(Y\)** that is explained by \(x\), the independent variable. This tells us how much of the variation in \(Y\) can be attributed to \(x\), without considering any model fitting.

### Summary:
- The first expression (\(1 - \frac{\text{RSS}}{\text{TSS}}\)) is equivalent to **R-squared**, which represents the proportion of the variation in the outcome variable \(Y\) that is explained by the model.
- `fitted_model.rsquared` is simply the R-squared value computed by the model and quantifies its accuracy in explaining \(Y\).
- `np.corrcoef(Y, fitted_model.fittedvalues)[0,1]**2` and `np.corrcoef(Y, x)[0,1]**2` are both capturing the **proportion of variation** explained by the relationship between the variables, where the first is for the model's predictions, and the second is for the raw predictor variable \(x\).

````python
#1. Bootstrapping the Fitted Slope Coefficients
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf
import plotly.express as px

# Define the long wait times
long_wait_limit = 71
long_wait = old_faithful.waiting > long_wait_limit

# Get the long wait data
long_wait_data = old_faithful[long_wait]

# Bootstrap process
n_bootstrap = 1000
bootstrap_slopes = []

for _ in range(n_bootstrap):
    bootstrap_sample = long_wait_data.sample(n=len(long_wait_data), replace=True)
    model = smf.ols('duration ~ waiting', data=bootstrap_sample).fit()
    bootstrap_slopes.append(model.params['waiting'])

# Convert bootstrap slopes to a DataFrame
bootstrap_slopes_df = pd.DataFrame(bootstrap_slopes, columns=['slope'])

# Visualize the distribution of the bootstrapped slope coefficients
fig = px.histogram(bootstrap_slopes_df, x='slope', title='Bootstrapped Sampling Distribution of Slope Coefficients')
fig.show(renderer="png")
````

2. Simulating Data Under the Null Hypothesis (No Linear Association)

````python
# Define the null hypothesis parameters
beta_0 = 1.65
beta_1 = 0  # Null hypothesis: no linear relationship
sigma = 0.37

# Simulate samples under the null hypothesis
n_simulations = 1000
simulated_slopes = []
x_values = long_wait_data['waiting'].values

for _ in range(n_simulations):
    # Simulate y values (Y = beta_0 + beta_1*X + error)
    errors = np.random.normal(loc=0, scale=sigma, size=len(x_values))
    y_simulated = beta_0 + beta_1 * x_values + errors
    
    # Fit linear model to simulated data
    simulated_data = pd.DataFrame({'waiting': x_values, 'duration': y_simulated})
    model = smf.ols('duration ~ waiting', data=simulated_data).fit()
    simulated_slopes.append(model.params['waiting'])

# Convert simulated slopes to a DataFrame
simulated_slopes_df = pd.DataFrame(simulated_slopes, columns=['slope'])

# Visualize the sampling distribution of the simulated slopes
fig = px.histogram(simulated_slopes_df, x='slope', title='Sampling Distribution of Slope Coefficients (Null Hypothesis)')
fig.show(renderer="png")
````

3. Bootstrapped Confidence Interval and p-value Comparison

````python
# Calculate the 95% bootstrapped confidence interval
bootstrap_ci = np.percentile(bootstrap_slopes, [2.5, 97.5])

# Check if 0 is within the confidence interval
contains_zero = 0 >= bootstrap_ci[0] and 0 <= bootstrap_ci[1]

# Simulate p-value based on the null hypothesis distribution
real_slope = smf.ols('duration ~ waiting', data=long_wait_data).fit().params['waiting']
simulated_p_values = np.mean(np.array(simulated_slopes) >= real_slope)

# Print results
print(f"Bootstrapped 95% Confidence Interval for Slope: {bootstrap_ci}")
print(f"Does 0 lie within the 95% Confidence Interval? {contains_zero}")
print(f"Simulated p-value: {simulated_p_values}")

# Compare with real p-value from the actual dataset
real_p_value = smf.ols('duration ~ waiting', data=long_wait_data).fit().pvalues['waiting']
print(f"Real p-value from the actual dataset: {real_p_value}")
````

2. **Simulating Data:**

3. **Confidence Interval and p-value:**

### (A) **Permutation Test**

A **permutation test** is a non-parametric method for testing hypotheses about the equality of two groups. It works by testing the null hypothesis \( H_0 \) that there is no difference between the groups (i.e., the means of the two populations are the same). The test involves shuffling the group labels and comparing the observed difference in means to the distribution of differences in means obtained from these label shufflings.

**Steps to Perform a Permutation Test**:
1. **Null Hypothesis (\( H_0 \))**: Assume that the two groups (short wait and long wait) are exchangeable under the null hypothesis (i.e., there is no difference in the durations between the two groups).
2. **Compute the Observed Difference in Means**: Calculate the difference in means between the "short" and "long" groups from the original dataset.
3. **Shuffle the Group Labels**: Randomly shuffle the group labels (i.e., assign the "short" and "long" labels to the data points at random).
4. **Compute the Difference in Means for Each Shuffling**: After shuffling, calculate the difference in means between the two groups for each iteration.
5. **Repeat the Process**: Repeat steps 3 and 4 a large number of times (e.g., 10,000 permutations) to create a distribution of differences under the null hypothesis.
6. **Compare the Observed Difference to the Distribution**: Compare the observed difference in means to the distribution of differences from the shuffling. The p-value is the proportion of permutations where the absolute difference in means is as extreme as or more extreme than the observed difference.

This approach does not make assumptions about the distribution of the data, making it a flexible, non-parametric test.

### (B) **Bootstrap Confidence Interval**

A **bootstrap confidence interval** is another non-parametric approach, but it focuses on estimating the sampling distribution of the statistic (in this case, the difference in means) by resampling with replacement from the original sample.

**Steps to Perform a Bootstrap Confidence Interval**:
1. **Resample the Data**: Create many bootstrap samples (e.g., 10,000) by randomly resampling with replacement from the "short" and "long" groups.
2. **Calculate the Mean Difference**: For each bootstrap sample, calculate the difference in means between the "short" and "long" groups.
3. **Repeat the Process**: Repeat the resampling and calculation of the mean difference multiple times (e.g., 10,000 resamples) to generate a distribution of mean differences.
4. **Construct the Confidence Interval**: From the distribution of bootstrap differences, calculate the 2.5th and 97.5th percentiles (or other percentiles depending on the desired confidence level) to create the 95% bootstrap confidence interval.

This method is computationally intensive but provides an empirical way to estimate the uncertainty of the difference in means.

### (a) **Explanation of the Sampling Approaches for the Two Simulations**

- **Permutation Test**: This method tests the null hypothesis by assessing how likely the observed difference in means is under the assumption that there is no real difference between the groups. By shuffling the group labels, we create a distribution of mean differences that would occur if the null hypothesis were true. This is a direct test of the null hypothesis and does not rely on assumptions about the distribution of the data.
  
- **Bootstrap Confidence Interval**: This method estimates the uncertainty in the difference of means by resampling from the observed data. It doesn't assume any specific distribution and focuses on the variability of the statistic (mean difference) itself. By generating a range of possible mean differences (based on resampling), it quantifies how much the mean difference might vary if the experiment were repeated many times.

Both methods are resampling-based and do not rely on traditional parametric assumptions, but they differ in their goal:
- The **permutation test** tests a specific hypothesis (null hypothesis of no difference).
- The **bootstrap confidence interval** estimates the uncertainty of the mean difference.

**Differences**:

The **permutation test** and **bootstrap confidence interval**, on the other hand, make fewer assumptions and are non-parametric methods.

### Ethical Professionalism Considerations
    
The TUT and HW both addressed some of the assumptions used in **Simple Linear Regression**. The **p-values** provided by `statsmodels` via `smf.ols(...).fit()` depend on these assumptions, so if they are not (at least approximately) correct, the **p-values** (and any subsequent claims regarding the "evidience against" the **null hypothesis**) are not reliable. In light of this consideration, describe how you could diagnostically check the first three assumptions (given below) when using analyses based on **Simple Linear regression** model. From an Ethical and Professional perspective, do you think doing diagnostic checks on the assumptions of a **Simple Linear regression** model is something you can and should do whenever you're doing this kind of analysis? 
            
> The first three assumptions associated with the **Simple Linear regression** model are that
> 
> - the $\epsilon_i$ **errors** (sometimes referred to as the **noise**) are **normally distributed**
> - the $\epsilon_i$ **errors** are **homoscedastic** (so their distributional variance $\sigma^2$ does not change as a function of $x_i$)
> - the linear form is [at least reasonably approximately] "true" (in the sense that the above two remain [at least reasonably approximately] "true") so that then behavior of the $Y_i$ **outcomes** are represented/determined on average by the **linear equation**)<br>
> 
>    and there are additional assumptions; but, a deeper reflection on these is "beyond the scope" of STA130; nonetheless, they are that<br><br>
> - the $x_i$ **predictor variable** is **measured without error**
> - and the $\epsilon_i$ **errors** are **statistically independent** (so their values do not depend on each other)
> - and the $\epsilon_i$ **errors** are **unbiased** relative to the **expected value** of **outcome** $E[Y_i|x_i]=\beta_0 + \beta_1x_i$ (which is equivalently stated by saying that the mean of the **error distribution** is $0$, or again equivalently, that the **expected value** of the **errors** $E[\epsilon_i] = 0$)
